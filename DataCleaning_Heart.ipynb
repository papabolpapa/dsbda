```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
heart=pd.read_csv('/content/heart.csv')
```

```python
#Data Cleaning
heart.fillna(heart.mode().iloc(0),inplace=True)
```

```python
heart.drop_duplicates(inplace=True)
print("Data Cleaning is done")
```

```python
#Data Integration
heart.columns
```

```python
df1=heart[['age','sex','cp','trestbps']]
df2=heart[['trestbps','chol','fbs','restecg']]
```

```python
#Data transformation
heart=heart.astype(str)
```

```python
encoder=LabelEncoder()
for col in heart.select_dtypes(include=['object']):
  heart[col]=encoder.fit_transform(heart[col])
```

```python
heart.head()
```

```python
#Model BUilding
x=heart.drop(columns=['age'])
y=heart['cp']
```

```python
y=y.round().astype(int)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```
